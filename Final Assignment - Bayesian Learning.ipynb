## The Consumer Problem

Each consumer (below I omit the index $i$ for consumers, since here all consumers have the same problem) is choosing a sequence of purchases to maximize the expected discounted sum of utility flows in all future periods:
\begin{align*}
    \max_{D=\{\{d_{jt}\}_{j=1}^{J}\}_{t=0}^{\infty}}{\mathbb{E}_D\left[\sum_{t=0}^{\infty}\beta^{t}\sum_{j=0}^{J}d_{jt}\left(u_{jt}+\epsilon_{jt}\right)\Big|x_0\right]}
\end{align*}

where $u_{jt}:=\gamma-\exp(-\rho\xi_{jt})-\alpha p_{jt}$ is the utility from product $j$ given price $p_{jt}$ and perceived quality $\xi_{jt}$ at $t$ ($\gamma,\rho,\alpha$ are parameters to be explained below); the utility from product $j$ also has an i.i.d. (accross products and time) random shock observed to the consumer but not to the researcher $\epsilon_{jt}$; $d_{jt}$ is an indicator that equals 1 if $j$ was consumed at $t$, and 0 otherwise; $\beta \in[0,1)$ is the discount factor and $x_0$ is the initial state.

## Match-Values, Learning and Uncertainty

In the problem analyzed here, there are $J$ products and one outside option $j=0$. Each product $j=1,...,J$ is characterized by a mean match-value $\bar{\mu}_j$. The match-value represents the quality of a given product, and it can differ accross consumers. In this case, each consumer $i$ has a specific match-value $\mu_j$ drawn from the distribution $N(\bar{\mu}_j,\tau^2)$. In other words, it is as if consumers had heterogeneous preferences for products, with the heterogeneity distributed normally in the population, around a mean match-value.

But consumers are uncertain about their true match-values $\mu_j$. They start with a prior belief and then update this belief using bayes rule after each consumption. Here, we assume that consumers have rational expectations and their initial prior is that $\mu_{j} \sim N(\bar{\mu}_j,\tau^2)$ (i.e., they expect their match values to be drawn from the population distribution, as they are in fact). This is represented in the initial state $x_0:=(...,\mu_{j0}=\bar{\mu}_j,...,\sigma_{j0}^2=\tau^2,...)$ in the problem above. It includes the mean and variances for each product $j=1,...,J$ at period $0$.

The consumer can obtain more information about its match-value for each product by consuming it, but the information is not perfect. For each $j$ and $t$, the consumer perceives the quality of the product as $\xi_{jt}=\mu_j+\nu_{jt}$, with $\nu_{jt} \sim N(0,\sigma_{\nu}^{2})$. The perceived quality is what enters the expected utility from each $j$ at each $t$.

Given the perceived quality from a product, the consumer then uses Bayes rule to update its prior beliefs about the match value. Because here we assume that both priors and the perceived quality have normal distributions, we know by conjugacy that the posterior distribution will also be normal, and we can just update its parameters using the closed-form expressions for this case. For a prior $x_t= (...,\mu_{jt},...,\sigma_{jt}^2,...)$ at period $t$, and perceived quality $\xi_{jt}$ (that is, only product $j$ was consumed at $t$):
\begin{align*}
    x_{t+1}=
    (...,\frac{\sigma_{jt}^2\xi_{jt}+\sigma^2_{\nu}\mu_{jt}}{\sigma^2_{\nu}+\sigma_{jt}^2},...,\frac{\sigma^2_{\nu} \sigma_{jt}^2}{\sigma^2_{\nu}+\sigma_{jt}^2},...)
\end{align*}

where the other components of $x_{t+1}$, related to products $k\ne j$, are unchanged.

## Induced Transition Density for Beliefs

Notice that each sequence of purchases $D$ induces a transition density on the state variables - which we call $f(x_{t+1}|x_t,d_t)$ - because consumers update their prior beliefs differently depending on the product consumed and on the random perceived quality; and the randomness in the perceived quality makes next-period's state random even given a product choice this period. Using the bayesian updating process assumed above, we can construct this transition density.

First, given a choice in period $t$, the transition of the mean and variance paramaters for other products not choosen is deterministic: they do not change. Also, the transition for the variance of product $j$, $\sigma_{jt}^2$, is deterministic, because it does not depend on the random perceived quality. Therefore, even though $f(x_{t+1}|x_t,d_t)$ is in principle a multivariate density, we can treat it as an univariate density for the element $\mu_{j,t+1}$ associated with the $j$ choosen at $t$, since all points for which the deterministic elements differ from bayesian updating have zero density. This gives a transition density:
\begin{align*}
    f(x_{j,t+1}|x_{jt},d_{jt}=1)=\begin{cases}
    N\Big(\mathbb{E}\left[\frac{\sigma_{jt}^2\xi_{jt}+\sigma^2_{\nu}\mu_{jt}}{\sigma^2_{\nu}+\sigma_{jt}^2}\Big|x_t\right],&Var\left[\frac{\sigma_{jt}^2\xi_{jt}+\sigma^2_{\nu}\mu_{jt}}{\sigma^2_{\nu}+\sigma_{jt}^2}\Big|x_t\right]\Big)=N(\mu_{jt},\frac{\sigma_{jt}^4}{\sigma^2_{\nu}+\sigma_{jt}^2}),\\
    &\text{ if $k \ne j$ as described by Bayes updating}\\
    0,\text{ otherwise}&
\end{cases}
\end{align*}

## Recursive Formulation of the Problem

The transition density defined above is markov, since it depends only on current state variables. We also assume the consumer knows this transition density (that is, the consumer knows she is bayesian and can use this to anticipate how her beliefs will evolve, given potential information to be received after each possible consumption choice). Given the stationarity of the utility function (and transition density), we can re-write the consumer problem recursively:
\begin{align*}
    V(x)&=\sum_{k=1}^{K}\omega_k\left(\int\max_{j}\left\{V_j(p_k,x)+\epsilon_j\right\}g(\epsilon)d\epsilon\right)\\
    &=\sum_{k=1}^{K}\omega_k\left(\int\max_{j}\left\{\mathbb{E}\left[u_j(p_k,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'+\epsilon_{j}\right\}g(\epsilon)d\epsilon\right)
\end{align*}

where $V_j(p,x):=\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'$ is the choice specific value function.

In words, each period, the consumer decides which product to purchase based on (i) the expected current period utility, given his current beliefs, prices and the $\epsilon$ shock (all known before consumption, and thus, in principle, state variables) - this is the first term inside the integral in the second line; (ii) and expected continuation value, given the transition density for beliefs (and the other state variables), conditional on current state and consumption choice - this is the second term inside the integral in the second line above. Also, above, we integrate out prices and the $\epsilon$ shock, obtaining an integrated Bellman equation for the integrated value function $V(x)$, which has only $x$ as state variable. Below I will explain in more detail why this integration of prices and shocks is possible.

Also, notice that the consumer considers expected utility, because consumption is decided prior to the realization of perceived quality. The parameter $\rho$ thus measures risk aversion, since it measures the rate at which utility decreases with higher uncertainty over match-value for a product (and $\alpha$ is price sensitivity):
\begin{align*}
    \mathbb{E}\left[u_{j}(p,x)|x=(\mu_j,\sigma_j^2)\right]=\gamma-\exp\left(-\rho\mu_{j}+\frac{1}{2}\rho^2(\sigma_{\nu}^2+\sigma_{j}^2)\right)-\alpha p_{j}
\end{align*}

We can then see from the expressions above that, given the integrated value function $V(x)$, we can calculate the choice-specific value functions $V_j(p,x)$ for given $(p,x)$ easily, using the functional form for utility. Then, because we assume $\epsilon$ are i.i.d. logit, centered at zero, we can also obtain conditional choice probabilities (CCPs) using the logit expression:
\begin{align*}
    \chi_j(p,x)=\frac{\exp\{V_j(p,x)\}}{\sum_{k=0,...,J}\exp\{V_k(p,x)\}}
\end{align*}

## Integration of Prices and $\epsilon$ Shocks

Even though we write the choice-specific value functions using both $p$ and $x$ as arguments (because consumers observe prices before choosing), the solution method reduces the state space by integrating over prices (and shocks) using the integrated value function, as already mentioned above. Now I will show the steps and assumptions used to obtain the integrated value function.

Before all integrations, we can write the Bellman equation generically as follows, using both $x,p,\epsilon$ as state variables and a generic transition function:
\begin{align*}
    V(x,\epsilon,p) = \max_{j=0,...,J}{\left\{\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int{V(x',\epsilon',p')h(x',\epsilon',p'|x,\epsilon,p,j)d(x',\epsilon',p')+\epsilon_j}\right\}}
\end{align*}

where $h(x',\epsilon',p'|x,\epsilon,p,j)$ is the generic transition density for the three state variables here.

However, we have: (i) $\epsilon$ is i.i.d. accross time and independent of $x'$ conditional on current state and choice $(x,j)$; (ii) $p$ is i.i.d. accross time and also independent of $x'$ conditional on current state and choice $(x,j)$. Also assume $\epsilon$ and $p$ are independent. These assumptions allow us to rewrite the transition density as the multiplication of three different densities: $h(x',\epsilon',p'|x,\epsilon,p,j)=f(x'|x,j)g(\epsilon')\omega(p')$. Then we can re-write the bellman:
\begin{align*}
    V(x,\epsilon,p) &= \max_{j=0,...,J}\left\{\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'+\epsilon_{j}\right\}
\end{align*}

where $V(x'):=\sum_{k=1}^{K}\omega_k\left(\int V(x',\epsilon',p_k)g(\epsilon')d\epsilon'\right)=\sum_{k=1}^{K}\omega_k\left(\int \max_{j}\{V_j(p_k,x')+\epsilon'\} g(\epsilon')d\epsilon'\right)$ - that is, the integrated value function we defined above.

Then we can integrate both sides over prices and shocks to get the integrated Bellman equation we showed above:
\begin{align*}
    V(x)&=\sum_{k=1}^{K}\omega_k\left(\int  V(x,\epsilon,p_k) g(\epsilon)d\epsilon\right) =\sum_{k=1}^{K}\omega_k\left( \int \max_{j=0,...,J}\left\{\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'+\epsilon_{j}\right\}g(\epsilon)d\epsilon\right)
\end{align*}

This is a Bellman equation on the integrated value function $V(x)$, and satisfies the contraction mapping properties, so that we can obtain the integrated value function for this problem by iterating on this Bellman from an initial guess. Then we plug the obtained $V(x)$ on the expression for the choice specific value function to get the value functions for each $(p,x)$ pair, and then obtain the CCPs through the logit expression shown above.

Therefore, we need prices to be i.i.d., conditionally independent from $x$ and independent from $\epsilon$ in order to be able to rewrite the transition density is a separable way, and then integrate out prices, leaving us only with $x$ as the state variable in the integrated value function as we defined (and similarly for $\epsilon$). 

Intuitively, all information needed to make a decision each period is contained in current state variables, because the markov structure of the transition function implies that current state variables and decision is all we need to know to assess the impact of the current decision on the future value of the problem. 

However, the i.i.d. structure of prices and the $\epsilon$ shock implies that their current values do not help in predicting their future values; and the conditional independence of $\epsilon$ and $p$ imply that their current values do not help predict future values of the $x$ state variable. Thus, we do not need to carry that information to obtain the optimal decisions.

Notice that conditional independence of prices means that current prices only affect future belifes $x'$ through their impact on today's purchase decision $j$, but not directly. This rules out, for instance, current prices signaling quality, in which case I might update even beliefs about products that I did not purchase, because I saw their prices. If this assumption is violated, then we need to carry $p$ as a state variable and deal with a multivariate transition density in both $(x,p)$.

Also, the i.i.d. rules out that prices might show some inertia, due to several factors, ranging from cost factors to price stickness, for instance. If prices are not i.i.d., then when we integrate with respect to $p$ above, we would be using a conditional density $\omega(p'|p)$ (assuming we still have the conditional independence, and thus can separate the transition density), and the resulting integral would thus still be a function of current $p$. Therefore, we would not be able to eliminate $p$ as a state variable. Because $p$ is itself a vector, this would add $J$ state variables to the state vector $x$ (that already has $J$ means and variances), significantly increasing the computational burden. This is the main advantage of having an i.i.d. pricing process.

## Solution to the model

The strategy for solving the model above can be summarized in the following steps:

1. Guess initial chebyshev weights $\theta^0$. Since $V(x)$ is infinite-dimensional, and thus we cannot evaluate it at all possible states, we will approximate it using chebyshev polynomials. Choosing $\theta^0$ is thus akin to choosing the initial guess $V^0(x)$.

2. We need to calculate the RHS of the integrated Bellman equation:
\begin{align*}
    V(x)&=\sum_{k=1}^{K}\omega_k\left( \int \max_{j=0,...,J}\left\{\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'+\epsilon_{j}\right\}g(\epsilon)d\epsilon\right)\\
    &=\sum_{k=1}^{K}\omega_k\left( \log \sum_{j}\exp\left\{\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'\right\}\right)\\
\end{align*}

where we use in the second line the expression for the maximum of a sum of logit terms centered at zero. Since the RHS is a function of $x$, we evaluate it at the nodes used in the Chebyshev approximation. Thus, we pre-compute these nodes and chebyshev polynomials evaluated at these nodes to later use when calculating the RHS of the Bellman.

3. For each node $x$ and choice $j$, we need to calculate $\int V(x')f(x'|x,j)dx'$. To do this, we use Gauss-Hermite numerical integration. Because both Gauss-Hermite and Chebyshev procedures involve summations, we can obtain:

\begin{align*}
    \int V(x')f(x'|x,j)dx'&=\int \left[\sum_{k}\theta^0_k T_k(x')\right]f(x'|x,j)dx'\\
    &=\sum_{q}w_q \left[\sum_{k}\theta^0_k T_k(x'_q)\right]\\
    &=\sum_{k}\theta^0_k\left[\sum_{q}w_qT_k(x'_q)\right]\\
    &=\sum_{k}\theta^0_k\mathbb{E}_k(x,j)\\
\end{align*}

where $\mathbb{E}_k(x,j)=\int T_k(x')f(x'|x,j)dx'$ is the expectation of the $k$-th chebyshev polynomial using the transition density. Thus, for each point $(x,j)$, we pre-calculate the inside expectation $\mathbb{E}_k(x,j)$, and then just call these values when calculating the integral above for each new $\theta$.

4. Now, for each node $x$, choice $j$ and price vector $p$ we pre-calculate $\mathbb{E}\left[u_j(p,x)|x\right]=\gamma-\exp\left(-\rho\mu_{j}+\frac{1}{2}\rho^2(\sigma_{\nu}^2+\sigma_{j}^2)\right)-\alpha p_{j}$.

5. Using the pre-calculations above and the current guess for $\theta$, we get, for each tuple $(x,j,p)$, the choice-specific value function $V_j(p,x)$. Then, using the logit expression and the distribution of prices, we can get updated values for $V^1(x)$ for each node point, as shown by the expression in step 2 above.

6. Then, finally, we get $\theta^1$ from the regression $V^1(x)=T\theta$, whete $T$ are the chebyshev polynomials pre-calculated in step 1 above and evaluated at the node points. Then we can calculate $\max|\theta^0-\theta^1|$ and, if small enough, stop; or, if not small enough, use $\theta^1$ as the new initial guess and start again from step 5 above.

### Chebyshev Approximation

To pre-calculate the nodes and chebyshev polynomials, I first implement two auxiliary functions:

1. *expandgrid*: this function takes a vector of nodes (for one dimension only and still in the interval $[-1,1]$) and the dimension $D$ of the function to be approximated and creates the $D-$fold cartesian product of the nodes. The output is a matrix of size $chebM^{D}\times D$, where $chebM$ is the number of nodes used in each dimension for the approximation.

```julia
function expandgrid(nodes,D)
    temp = nodes
    for d in 1:(D-1)
        ret = Any[]
        for i in nodes
            for r in temp
                push!(ret, vcat(i,r))
            end
        end
        temp = ret
    end
    temp = vcat([x' for x in temp]...)
    return temp
end
```

2. *calculateChebyshevPolynomials*: this function takes a vector of nodes (for one dimension only and still in the interval $[-1,1]$) and the highest degree of the polynomials to be used, $chebN$, and returns a matrix of size $chebM \times (chebN+1)$. The element in row $m$ and column $n$ is the $n-$degree chebyshev polynomial applied to the $m$ node, which are calculated in a loop, using the recursive expression for the polynomials.

```julia
function calculateChebyshevPolynomials(nodes,chebN)
    T = fill(1.0,size(nodes)[1],chebN+1)
    if chebN >= 1
        T[:,2] .= nodes
        if chebN >= 2
            for i in 2:chebN
                T[:,i+1] .= 2.0 .* nodes .* T[:,i] .- T[:,i-1]
            end
        end
    end
    return(T)
end
```

The function *initChebyshev* returns the relevant pre-calculated matrices to perform chebyshev approximation.

The function takes as arguments: two vectors of size $D$ containing the upper and lower bounds for the domain in each dimension of the function to be approximated ($chebUpperBnd, chebLowerBnd$); the degree and number of nodes for the approximation; and dimension of the function.

It first calculates the values for the nodes on $[-1,1]$ and creates the $D$-fold cartesian product using *expandgrid*, resulting in the matrix $X$. Then it rescales the elements in $X$ to the original intervals, creating the matrix $Z$.

After that, it calls *calculateChebyshevPolynomials* to calculate the polynomials for each original node, and then uses kronecker products to combine the one dimension polynomials and obtain all the polynomials in the $D$-fold tensor product. The result is a big matrix, $bigT$, of size $chebM^D \times (chebN+1)^D$. Each element in this matrix is a (multidimensional) chebyshev polynomial (indexed by the columns) applied to a (multidimensional) node (indexed by rows). Because the chebyshev polynomials are product combinations of the single dimensional polynomials, we have $(chebN+1)^D$ columns; and similarly, we have $chebM^D$ rows as well.

Lastly, the function already pre-computes the standard matrix used in OLS regressions, $A$, to use when we have the updated value functions in order to obtain the updated chebyshev coefficients.

```julia
function initChebyshev(chebUpperBnd, chebLowerBnd, chebN, chebM, D)    
    nodes = -cos.(pi .* ( ( 2.0 .* [1.0:chebM;] .- 1.0 )./(2.0*chebM) ))
    X::Array{Float64,2} = expandgrid(nodes,D)
    Z = ( (X .+ 1.0).*(chebUpperBnd .- chebLowerBnd).*0.5 ) .+ chebLowerBnd
    
    T = calculateChebyshevPolynomials(nodes,chebN)    
    bigT = T
    for d in 2:D
        bigT = kron(bigT,T)
    end
    
    A = (bigT'*bigT)\bigT'
    
    return(X = X, Z = Z, A = A)
end
```

### Set Problem

The function *SetProblem* takes the main paramaters of the model and pre-calculates some objects as described in steps 1-4 outlined above. The function does the following:

1. Uses the *gausshermite* function from the *FastGaussQuadrature* package to compute nodes and weights for a single dimensional Guass-Hermite integration of degree $quadN$. It results in two vectors of size $quadM$, containing the integration nodes and weights. Although the value function is multivariate, below I show that in the integration step, we actually only need to do univariate integration, and thus these single dimensional integration nodes suffice.

2. Call *initChebyshev* to obtain the relevant objects for Chebyshev approximation, as described above.

3. Call *preComputeChoiceValues* to pre-compute, for each pair $(x,j)$, the expected values: $\mathbb{E}_k(x,j)=\int T_k(x')f(x'|x,j)dx'\approx\sum_{q}w_qT_k(x'_q)$ using the Gauss-Hermite nodes and weights:

    a. Here, we need to calculate $\mathbb{E}_k(x,j)$ for each state in the chebyshev nodes, which are the $chebM^D$ rows in the matrix $Z$. As a first step, for a given state $s$ in the loop, we recover the corresponding row of $Z$, which has the current means of the match-value prior as the first $J$ elements, and the variances as the last $J$ elements. We then update it to reflect the next state ($s\_next$), since the integral we are interested in is over the next period's possible states, given potential information received today.
    
    b. For a given product $j$ choosen, we can then readily update the corresponding variance on $s\_next$ using the expression shown above in question 1: $\sigma^2_{j,t+1}=\frac{\sigma^2_{\nu} \sigma_{jt}^2}{\sigma^2_{\nu}+\sigma_{jt}^2}$.
    
    c. Now, next state prior mean for product $j$ is random before consumption is actually done. But we know its distribution, given current state, as shown in question 1; and we need to calculate the integrals over the chebyshev polynomials using this distribution.
    
    As a first step we repeat the next state vector as rows in a matrix of size $quadM \times D$, since we need to evaluate the chebyshev polynomials at each Gauss-Hermite node to perform the integration.
    
    Then we use the change of variables expression for Gauss-Hermite integration to get the "updated" value of each node to be used in the integration: $\mu_{jt}+x_q\sqrt{2\frac{\sigma_{jt}^4}{\sigma^2_{\nu}+\sigma_{jt}^2}}$, where $x_q$ is the original Gauss-Hermite node.
    
    Then we rescale the next state vector to $[-1,1]^D$, since our chebyshev polynomials have this as domain, and call the function *evalChebyshevPolyn* to get the polynomials evaluated at the relevant sates.
    
    Notice that, here, we want to calculate $\int T_k(x')f(x'|x,j)dx'$ for each chebyshev polynomial $k$. In principle, this is a multivariate integral. But we know that all elements of $x'$, except the mean for choosen product $j$, are deterministic. Thus, we can rewrite the integral as $\int T^j_k(x_j')\tilde{f}(x_j'|x,j)dx_j'$, where $T^j_k(x_j')$ is $T_k(x')$ applied at $x'$ with $j-$element equal to $x_j'$ and with all other elements fixed at their deterministic next period values. This integral is now single-dimensional, and we can approximate it using Gauss-Quadrature as well: $\sum_{q}w_qT^j_k(x'_{jq})$. Since multivariate chebyshev polynomials are just products of the single dimensional ones applied at elements of the node vector, $T^j_k(x'_{jq})$ can also be calculated as the product of the single-dimensional chebyshev polynomials applied on the single elements of the $s\_next$ vector, which gives the same $T_k(x'_{q})$ polynomial, with $x'$ having its $j$ product means and variances updated.
    
    That is what I do below. (i) First I calculate single-dimensional chebyshev polynomials for the $quadM$ nodes at dimension $d=1$, resulting in a matrix $quadM \times (chebN+1)$; and similarly for dimension $d=2$. (ii) Then, for each node, I do the kronecker product of all single-dimensional polynomials applied at those nodes for both states, obtaining their product combinations, saving the result in a $chebM \times (chebN+1)^2$ matrix. (iii) Using this resulting matrix, I proceed, now calculating the chebyshev polynomials for the nodes at dimension $d=3$, and doing a kronecker product with the combinations of dimensions 1 and 2. (iv) I do this until I obtain a matrix of dimension $quadM \times (chebN+1)^D$, which contains, for each integration node, all multivariate chebyshev polynomials in its columns (resulting from product combinations of the single-dimensional polynomials evaluated at the relevant next period's states).
    
    Lastly, I use the Gauss-Hermite weights to do the integration, and get a matrix that has, in each $k$ row, $\mathbb{E}_k(x,j)$.
    
4. The last matrix to be pre-calculated contains expected utilities for each tuple of state-choice-prices, $(x,j,p)$, and follows the expression for the expected utility shown in question 1: $\mathbb{E}\left[u_{j}(p,x)|x=(\mu_j,\sigma_j^2)\right]=\gamma-\exp\left(-\rho\mu_{j}+\frac{1}{2}\rho^2(\sigma_{\nu}^2+\sigma_{j}^2)\right)-\alpha p_{j}$.


```julia
function evalChebyshevPolyn(s_next,chebN)
    rows = size(s_next)[1]
    D = size(s_next)[2]
    
    Tn0 = calculateChebyshevPolynomials(s_next[:,1],chebN)
    for i in 2:D
        Tn1 = calculateChebyshevPolynomials(s_next[:,i],chebN)
        Tnn = fill(0.0,rows,(chebN+1)^(i))
        for m in 1:rows
            Tnn[m,:] = kron(Tn0[m,:],Tn1[m,:])
        end
        Tn0 = Tnn
    end
    return(Tn0)
end
```

```julia
function preComputeChoiceValues(chebN,chebM,D,J,chebLowerBnd,chebUpperBnd,
        Z,sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,
        alpha,prices,rhoSignal,rhoRisk,n_states)
    
    E_k_xj = fill(0.0,(chebN+1)^D,J+1,n_states)
    
    # For each state-choice pair in Z
    for s in 1:n_states
        for j in 1:(J+1)
            
            if j==(J+1)
                #When no-purchase, state does not change
                s_next = Z[s,:]'
                s_next = (2.0 .* (s_next .- chebLowerBnd)./
                    (chebUpperBnd .- chebLowerBnd) ) .- 1.0
                E_k_xj[:,j,s] = evalChebyshevPolyn(s_next,chebN)
            elseif Z[s,J+j]==0.0
                #When no-purchase, state does not change
                s_next = Z[s,:]'
                s_next = (2.0 .* (s_next .- chebLowerBnd)./
                        (chebUpperBnd .- chebLowerBnd) ) .- 1.0
                E_k_xj[:,j,s] = evalChebyshevPolyn(s_next,chebN)
            else
                #Get rows of state values
                s_past = Z[s,:]
                s_next = Z[s,:]
            
                #Update column of variance for j (deterministic)
                s_next[J+j] = (sigma2_v*s_past[J+j])/(sigma2_v + s_past[J+j])
                
                #repeat s_next vector quadM times for integration
                s_next = repeat(s_next',quadM,1)
                
                #get nodes for the next state mean for j choosen (random)
                mu_var = (s_past[J+j]^2.0)/(sigma2_v + s_past[J+j])
                s_next[:,j] = sqrt(2.0*mu_var).*gh_nodes .+ s_past[j]
                
                #Rescale nodes to evaluate chebyshev polynomials at them
                s_next = (2.0 .* (s_next .- chebLowerBnd)./
                            (chebUpperBnd .- chebLowerBnd) ) .- 1.0
                Tn0 = evalChebyshevPolyn(s_next,chebN)
                
                #Use gh_weights to integrate the cheb poly and get expectation
                E_k_xj[:,j,s] = (Tn0'*gh_weights)./sqrt(pi)
            end                
        end
    end
    
    utility = fill(0.0,J+1,K,n_states)
    #For each state-product-price
    for s in 1:n_states
        for k in 1:K
            for j in 1:J
                #Compute expected utility
                utility[j,k,s] = gamma - alpha * prices[k,j] - 
                    exp(-rhoSignal * Z[s,j] + 0.5*(rhoRisk^2.0)*(sigma2_v+Z[s,J+j]))
            end
        end
    end
    utility[J+1,:,:] = fill(0.0,1,K,n_states)
    
    return(E_k_xj=E_k_xj,utility=utility)
end
```

```julia
function SetProblem(parameters)
    
    @unpack chebUpperBnd, chebLowerBnd, prices, alpha, rhoSignal, rhoRisk, J, K,
        sigma2_v, gamma, D, beta, omega, quadN, chebN, chebM = parameters
    
    #Create gauss-hermite nodes
    gh_nodes, gh_weights = gausshermite(quadN)
    quadM = size(gh_nodes)[1]
    
    #Initialize Chebyshev matrices
    @unpack X, Z, A = initChebyshev(chebUpperBnd, chebLowerBnd, chebN, chebM, D)
    
    #Pre-compute utilities and E_k(x,j)
    @unpack E_k_xj, utility = preComputeChoiceValues(chebN,chebM,D,J,chebLowerBnd,
        chebUpperBnd,Z,sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,alpha,prices,
        rhoSignal,rhoRisk,chebM^D)
        
    #Return pre-computed objects
    return(gh_nodes=gh_nodes, gh_weights=gh_weights, quadM=quadM,
        chebUpperBnd=chebUpperBnd, chebLowerBnd=chebLowerBnd, 
        chebN=chebN, chebM=chebM, D=D, X=X, Z=Z, A=A,
        E_k_xj=E_k_xj, utility=utility,prices=prices, alpha=alpha,
        rhoSignal=rhoSignal, rhoRisk=rhoRisk, J=J, omega=omega, K=K,
        sigma2_v=sigma2_v, gamma=gamma, beta=beta)
end
```

### Bellman operator

The function *bellmanOperator* uses the pre-calculated objects from *SetProblem* to iterate on the Bellman operator and get updated values for the chebyshev coefficients $\theta$.

It first calculates choice-specific value functions for every tuple $(x,j,p)$, using the expression: $V_j(p,x):=\mathbb{E}\left[u_j(p,x)|x\right]+\beta\int V(x')f(x'|x,j)dx'\approx \mathbb{E}\left[u_j(p,x)|x\right]+ \beta\sum_{k}\theta_k\mathbb{E}_k(x,j)$. Then it calculates the updated value function evaluated at the chebyshev nodes using the logit expression: $y(x) = \sum_{k=1}^{K}\omega_k\left( \log \sum_{j}\exp\left\{V_j(p,x)\right\}\right)$, and uses the chebyshev regression matrix $A$ to update the chebyshev coefficients.

```julia
function bellmanOperator(theta0,chebM,D,J,K,utility,beta,E_k_xj,omega,A)    
    choiceSpecValue = fill(0.0,J+1,K)
    y = fill(0.0,chebM^D,1)
    for s in 1:(chebM^D)
        temp = 0.0
        for k in 1:K
            for j in 1:(J+1)
                choiceSpecValue[j,k] = utility[j,k,s] +
                    beta*( dot(theta0,E_k_xj[:,j,s]) )
            end
            temp += omega[k]*( log( sum( exp.( choiceSpecValue[:,k]  ) ) )  )
        end
        y[s] = temp
    end
    theta1 = A*y
    return(theta1)
end
```

The function *solveProblem* below just iterates on the *bellmanOperator* in order to obtain the fixed point of the integrated bellman equation. The output is the vector of $\theta$ chebyshev coefficients that approximates the integrated value function $V(x)$.

```julia
function solveProblem(theta0,sp,error,tol)
    
    @unpack chebM,D,J,K,utility,beta,E_k_xj,omega,A = sp
    
    iter = 0
    theta1 = fill(0.0,size(theta0))
    while error > tol
        iter += 1
        theta1 = bellmanOperator(theta0,chebM,D,J,K,utility,beta,E_k_xj,omega,A)
        error = norm(theta1 .- theta0)
        theta0 = theta1
        if (iter%1000)==0
            @printf "Iteration: %d  Norm: %.15f\n" iter error
        end
    end
    
    return(theta1)
end
```

### Conditional choice probabilities 

The *CCP* function below computes conditional choice probabilities for given states. It takes as inputs a matrix in which each row is the state vector of prior means and variances, $x$, other problem information resulting from the *SetProblem* function and the true Chebyshev coefficients obtained after solving the problem. It returns a three-dimensional array with the choice probabilities for each product, for each state-price pair $(x,p)$. It also returns the choice specific value functions for the given states.

```julia
function CCP(theta_final,sp,states)
    
    @unpack chebN,chebM,D,J,chebLowerBnd,chebUpperBnd,
        sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,alpha,
        prices,rhoSignal,rhoRisk,beta = sp
    
    n_states = size(states)[1]
    
    @unpack E_k_xj, utility = preComputeChoiceValues(chebN,chebM,D,J,
        chebLowerBnd,chebUpperBnd,states,sigma2_v,quadM,gh_nodes,
        gh_weights,K,gamma,alpha,prices,rhoSignal,rhoRisk,n_states)
    
    CCP = zeros(J+1,K,n_states)
    trueChoiceSpecValue = zeros(J+1,K,n_states)
    for s in 1:(n_states)
        for k in 1:K
            for j in 1:(J+1)
                trueChoiceSpecValue[j,k,s] = utility[j,k,s] +
                    beta*( dot(theta_final,E_k_xj[:,j,s]) )
            end
            CCP[:,k,s] = ( exp.(trueChoiceSpecValue[:,k,s] ) ) ./
                sum( exp.(trueChoiceSpecValue[:,k,s] ) )
        end
    end
    return(probs=CCP,util=utility,Emax=E_k_xj,choiceSpecValue=trueChoiceSpecValue)
end
```

## b-d)

The function *simulationLoops*, starting from the initial prior, calculates the expected choice specific values (conditional on draws for shocks and prices) and updates the prior according to the choice decision - repeating this for $T$ periods; the whole process is repeated $N$ times and the averages for each $t$ give the choice probabilities (which are calculated in the function *computeFreq*). The function *simulateData* takes an initial prior and true match values, generates random numbers to obtain the values of $p$, $\epsilon$ and $\nu$ for each period $t=1,...,T$ and each $n=1,...,N$, and feed those to *simulationLoops*.

```julia
function simulationLoops(priceSequence,shockSequence,signalSequence,
        initPrior,theta_final,chebN,chebM,D,J,chebLowerBnd,
        chebUpperBnd,sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,
        alpha,prices,rhoSignal,rhoRisk,beta,T,N)
    
    expChoiceSpecValue = fill(0.0,J+1)
    choices = fill(0,T,N)
    aux = fill(0.0,D,T,N)
    histBeliefs = fill(0.0,D,T)
    
    for n in 1:N
        
        currentState = deepcopy(initPrior)
        
        for t in 1:T
            
            @unpack utility, E_k_xj = preComputeChoiceValues(chebN,chebM,D,J,
                chebLowerBnd,chebUpperBnd,currentState,sigma2_v,quadM,gh_nodes,
                gh_weights,K,gamma,alpha,prices,rhoSignal,rhoRisk,1)

            ind = 1
            for j in 1:(J+1)
                expChoiceSpecValue[j] = utility[j,priceSequence[t,n],1] + 
                    shockSequence[j,t,n] + beta*( dot(theta_final,E_k_xj[:,j,1]) )
            end
            
            max, ind = findmax(expChoiceSpecValue)
            choices[t,n] = ind
            if ind!=(J+1)
                currentState[ind] = (currentState[J+ind]*signalSequence[ind,t,n] +
                    sigma2_v*currentState[ind])/(currentState[J+ind] + sigma2_v)
                currentState[J+ind] = (sigma2_v*currentState[J+ind])/
                    (currentState[J+ind] + sigma2_v)
            end            
            
            aux[:,t,n] = currentState
            
        end
    
    end
    
    histBeliefs[:,:] = mean(aux,dims=3)
    return(choices=choices,histBeliefs=histBeliefs)    
end
```

```julia
function simulateData(theta_final,sp,initPrior,trueMatchValue;T=100,N=1000)
    
    @unpack chebN,chebM,D,J,chebLowerBnd,chebUpperBnd,
        sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,alpha,
        prices,rhoSignal,rhoRisk,beta,omega = sp
    
    cumOmega = [0.0 cumsum(omega,dims=2)]
    priceSequence = fill(0,T,N)
    temp = rand(Uniform(0,1),(T,N))
    for k in 1:K
        priceSequence[cumOmega[k] .< temp .<= cumOmega[k+1]]  .= k
    end
    
    shockSequence = rand(Gumbel(-0.57721,1),(J+1,T,N))
    
    signalSequence = repeat(trueMatchValue',1,T,N) .+ 
        ( randn(J,T,N).*sqrt(sigma2_v) )
    
    @unpack choices, histBeliefs = simulationLoops(priceSequence,shockSequence,
        signalSequence,initPrior,theta_final,chebN,chebM,D,J,chebLowerBnd,
        chebUpperBnd,sigma2_v,quadM,gh_nodes,gh_weights,K,gamma,alpha,prices,
        rhoSignal,rhoRisk,beta,T,N)
    
    return(choices=choices,histBeliefs=histBeliefs)    
end
```

```julia
function computeFreq(data)
    J = maximum(data)-1
    N = size(data)[2]
    T = size(data)[1]
    freq = fill(0.0,T,J+1)
    for j in 1:(J+1)
        temp = data .== j
        freq[:,j] = sum(temp,dims=2)./N
    end
    return(freq)
end
```

The function *switchFreq* below calculates the frequency of switches at every period of the simulated data generated above. 

Notice that I pick the $n$ runs that had an inside product choosen in the first period, and then use only those to calculate the evolution of switching probabilities. When I used all $N$ runs, the switching probabilities (even conditional on choosing an inside product) necessarily started very small, since, at period 2, it is the sum of the probabilities of choosing the sequences 1 and 2 or 2 and 1 in the first two periods, which are very small due to the fact the outside good has a much bigger choice probability. As $t$ grows, the values become higher, since there is a higher probability of an inside product being choosen in any of the previous periods. However, because I did not think that depicted well the experimentation going on in the beginning, I decided to use only runs where an inside product is picked in the first period.

```julia
function switchFreq(choices)
    ind = choices[1,:].!=3
    choice = choices[:,ind]
    T = size(choice)[1]
    N = size(choice)[2]
    switches = fill(0,T,N)
    for n in 1:N
        for t in 2:T
            if choice[t,n] != 3
                for i in ( (t-1):(-1):1 )
                    if choice[i,n] != 3
                        if choice[t,n]!=choice[i,n]
                            switches[t,n] = 1
                        end
                        break
                    end
                end
            end
        end
    end
    return(mean(switches,dims=2))    
end
```

### Simulations

The function *solveSimulate* below is just a convenient way of wrapping together everything we need to analyze the questions in the report under different parameters. It takes parameters as inputs and, possibly, values of the true chebyshev coefficients if the model was already solved - if not, then it solves the model. Then it runs the simulations and returns tables with choice probabilities, evolution of beliefs and switching probabilities conditional on choosing an inside product.

```julia
function solveSimulate(parameters,initPrior,trueMatchValue,
            T,N,theta_final,solve)
    
    Random.seed!(0)
    
    #Get parameters and solve model
    @unpack prices,K,omega,gamma,sigma2_v,alpha,rhoSignal,rhoRisk,
        chebUpperBnd,chebLowerBnd,chebN,chebM,D,J,beta = parameters
    
    sp = SetProblem(parameters)
    
    if solve
        error = Inf
        tol = 1e-10
        theta_final = solveProblem(theta_final,sp,error,tol)
    end
    
    #Get initial and complete information choice probs
    
    @show initPrior
    @show trueMatchValue
    
    @unpack probs, util, Emax = CCP(theta_final,sp,initPrior)
    initTotalProb = probs[:,1,:].*omega[1] .+ probs[:,2,:].*omega[2]
    @show initTotalProb
    
    @unpack probs, util, Emax = CCP(theta_final,sp,[trueMatchValue 0.0 0.0])
    complInfTotalProb = probs[:,1,:].*omega[1] .+ probs[:,2,:].*omega[2]
    @show complInfTotalProb
    
    #Evolution of Choice Probs
    
    @unpack choices,histBeliefs = simulateData(theta_final,sp,initPrior,
        trueMatchValue,T=T,N=N);
    freq = computeFreq(choices)
    
    #Evolution of Switching Prob
    
    switches = switchFreq(choices)
    ratio = switches./sum(freq[:,1:J],dims=2)
    
    return(choices=choices,histBeliefs=histBeliefs,freq=freq,
        switches=switches,ratio=ratio,theta_final=theta_final)
    
end
```
